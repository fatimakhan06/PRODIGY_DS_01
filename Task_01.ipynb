```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
# Filter data for the year 2022
df_2022 = df[df['Year'] == 2022]
```

```python
# Sort countries by population
df_2022_sorted = df_2022.sort_values(by='Value', ascending=False).head(10)
```

```python
#Graph 1: Bar Chart
plt.figure(figsize=(12, 6))
plt.bar(df_2022_sorted['Country Name'], df_2022_sorted['Value'], color='skyblue')
plt.title('Top 10 Countries by Population in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Graph 2: Line Chart
# Population growth for a specific country, e.g., India
india_data = df[df['Country Name'] == 'India']
years = india_data['Year']
population = india_data['Value']

plt.figure(figsize=(12, 6))
plt.plot(years, population, marker='o', linestyle='-', color='green')
plt.title('Population Growth of India (1960-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.tight_layout()
plt.show()
```

```python
# Graph 3: Histogram
# Plot the distribution of populations across all countries in 2022
plt.figure(figsize=(12, 6))
plt.hist(df_2022['Value']/1e6, bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Country Populations in 2022')
plt.xlabel('Population (Millions)')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
